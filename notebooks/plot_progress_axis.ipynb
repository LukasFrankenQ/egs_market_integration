```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
root = Path.cwd().parent.parent.parent / "cluster_data" / "breyer_sweep"
os.listdir(root)

template = "egs_costs_s_72_{}.csv"
years = list(range(1990, 2055, 5))
regions = pd.read_csv(root / template.format(2000))["name"].tolist()

mins = pd.DataFrame(index=regions, columns=years)
means = pd.DataFrame(index=regions, columns=years)

for year in years:

    costs = pd.read_csv(root / template.format(year)).set_index("name")

    mins[year] = costs["capex_min"]
    means[year] = costs["capex_mean"]
```

```python
for region in regions:
    if mins.loc[region].isna().any():
        continue

    fig, ax = plt.subplots()
    
    ax.plot(years, mins.loc[region], label=region)
    ax.plot(years, means.loc[region], label=region)

    plt.show()
```
